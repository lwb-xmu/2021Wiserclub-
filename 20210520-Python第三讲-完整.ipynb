```python
def fav_furong(n):
    zhushi_another_list = []
    xiaodian_another_list = []
    zuhshi_list = ['', '', '', '', '']
    xiaodian_list = ['', '', '', '', '']
    while len(zhushi_another_list) < n:
        zhushi = input('')
        if not "" in zhushi and zhushi not in zuhshi_list and zhushi not in zhushi_another_list:
            zhushi_another_list.append(zhushi)
    while len(xiaodian_another_list) < n:
        xiaodian = input('')
        if not "" in xiaodian and xiaodian not in xiaodian_list and xiaodian not in xiaodian_another_list:
            xiaodian_another_list.append(xiaodian)

    print("")
    print("")
    for zhushi in zhushi_another_list:
        print(zhushi)
    print("")
    print("")
    for xiaodian in xiaodian_another_list:
        print(xiaodian)
    return zhushi_another_list, xiaodian_another_list
```
